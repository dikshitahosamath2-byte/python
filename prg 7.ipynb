```python
def is_prime(num):
    if num <= 1:
        return False
    
    for i in range(2, num):
        if num % i == 0:
            return False
    
    return True
                                                                  # def keyword
def main():
    number = int(input("Enter a number: "))

    if is_prime(number):
        print(number, "is a Prime number")
    else:
        print(number, "is not a Prime number")

main()
```
