# set comprehension 

```python
s=[x for x in range(1,6)]
print(s)
```

```python
s=[x*x for x in range(1,6)]
print(s)
```

```python
squares = {x**2 for x in range(1, 11)}
print(squares)
```

```python
text = "hello world"
uppercase_letters = {char.upper() for char in text if char.isalpha()}
print(uppercase_letters)
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

common = {x for x in list1 if x in list2}
print(common)
```

```python
nums = {x for x in range(1, 101) if x % 3 == 0 and x % 5 == 0}
print(nums)
```

```python
words = ["apple", "banana", "cherry", "avocado"]
first_letters = {word[0] for word in words}
print(first_letters)
```

```python
s=[24,26,27,65]
s.remove(27)                                    #remove vs discard
print(s)
```

```python
s4={20,30,40,23,12,32}
s4.pop()                                             # pop vs clear
print(s4)                       
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print(A & B)                                              #intersection and union
print(A.intersection(B))
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print(A | B)
print(A.union(B))
```

```python
def even_odd():
    a = int(input("Enter a number: "))
    if a % 2 == 0:
        print(a, "is Even")
    else:
        print(a, "is Odd")

even_odd()
```

```python
def prime():
    a = int(input("Enter a number: "))
    if a > 1:
        for i in range(2, a):
            if a % i == 0:
                print(a, "is not Prime")
                break
        else:
            print(a, "is Prime")
    else:
        print(a, "is not Prime")

prime()
```
