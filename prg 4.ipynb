# write python script to print the first multiplies of N 

```python
# Get input from user
N = int(input("Enter a number: "))

# Print first 10 multiples of N
for i in range(1, 11):
    print(N * i)
```

# write python script to print the first M multiplies of N

```python
# Get inputs from user
N = int(input("Enter the number (N): "))
M = int(input("Enter how many multiples to print (M): "))

# Print first M multiples of N
for i in range(1, M + 1):
    print(N * i)
```

# write python script to print the first 10 multiplies of N in reverse order

```python
# Get input from user
N = int(input("Enter a number (N): "))

# Print first 10 multiples of N in reverse order
for i in range(10, 0, -1):
    print(N * i)
```

# write python script to print the table of users choice

```python
# Get number from user
num = int(input("Enter the number to print its table: "))

# Print multiplication table (1 to 10)
print(f"\nMultiplication Table of {num}:\n")

for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```
