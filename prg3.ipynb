# match stmts

```python
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 400:
            return "Bad Request"
        case 404:
            return "not found"
        case 500:
            return "internal server error"
        case _:
            return "unknown status"
print(http_status(500))
print(http_status(400))
```

# arithmetic operators

```python
def http_status(code):
    match code:
        case +:
            return "OK"
        case 400:
            return "Bad Request"
        case 404:
            return "not found"
        case 500:
            return "internal server error"
        case _:
            return "unknown status"
print(http_status(500))
print(http_status(400))
```

```python
def http_status(code):
    return (
        (code == 200) * "OK" +
        (code == 400) * "Bad Request" +
        (code == 404) * "not found" +
        (code == 500) * "internal server error" +
        (1 - ((code == 200) + (code == 400) + (code == 404) + (code == 500))) * "unknown status"
    )

print(http_status(500))
print(http_status(400))
```

```python
def arithmetic_operation(a, b, op):
    match op:
        case '+':
            return a + b
        case '-':
            return a - b
        case '*':
            return a * b
        case '/':
            if b != 0:
                return a / b
            else:
                return "Division by zero not allowed"
        case _:
            return "Invalid operation"

print(arithmetic_operation(10, 5, '+'))
print(arithmetic_operation(10, 5, '*'))
print(arithmetic_operation(10, 5, '-'))
print(arithmetic_operation(10, 5, '/'))
```
