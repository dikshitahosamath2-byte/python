# slicing operators

```python
l1=[12,34,56,34,66,67,68,76]
print(l1[0:7:3])
```

```python
l1=[12,34,56,34,66,67,68,76]
print(l1[1:5])
```

```python
l1=[12,34,56,34,66,67,68,76]
print(l1[:4])
print(l1[0:5])
print(l1[2:4])
print(l1[:4])
print(l1[5:])
print(l1[-3:])
print(l1[:-2])
print(l1[1:2])
```
