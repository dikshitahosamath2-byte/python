```python
import time
size=10_00_000
l1=list(range(size))
l2=list(range(size))
start=time.time()
l3=[x*y for x,y in zip(l1,l2)]
end=time.time()
print(l3[0:10])
print(f"Time taken by a list is {end-start}")
```

```python
import time
import numpy as np
size=10_00_000
arr1=np.array(list(range(size)))
arr2=np.array(list(range(size)))
start=time.time()
arr3=arr1*arr2
end=time.time()
print(arr3[0:10])
print(f"Time taken by a list is {end-start}")
```
